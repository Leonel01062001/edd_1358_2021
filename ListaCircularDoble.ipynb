```
class NodoDoble:
    def __init__( self , dato , siguiente = None, anterior = None):
        self.dato = dato
        self.siguiente = siguiente
        self.anterior = anterior
        
class DoubleCircularList:
	def __init__(self):
		self.__ref = None
		self.__size = 0

	def is_empty(self):
		return self.__size == 0

	def insert(self, value ):
		
		if self.__ref == None:
			self.__ref = NodoDoble(value)

		elif value >= self.__ref.dato and self.__ref.siguiente == None:
			self.__ref.siguiente = NodoDoble(value,self.__ref,self.__ref)
			self.__ref.anterior = self.__ref.siguiente
			self.__ref = self.__ref.siguiente

		elif value >= self.__ref.dato:
			self.__ref.siguiente = NodoDoble(value,self.__ref.siguiente,self.__ref)
			self.__ref.siguiente.siguiente.anterior = self.__ref.siguiente
			self.__ref = self.__ref.siguiente

		elif value <= self.__ref.dato and self.__ref.siguiente == None: 
			self.__ref.siguiente = NodoDoble(value,self.__ref,self.__ref)
			self.__ref.anterior = self.__ref.siguiente

		elif value <= self.__ref.siguiente.dato:
			self.__ref.siguiente = NodoDoble(value,self.__ref.siguiente,self.__ref)
			self.__ref.siguiente.siguiente.anterior = self.__ref.siguiente			

		else: 
			curr_node = self.__ref.siguiente
			while curr_node.dato != self.__ref.dato:
				if value > curr_node.dato and value < curr_node.siguiente.dato:
					curr_node.siguiente = NodoDoble(value,curr_node.siguiente,curr_node)
					curr_node.siguiente.siguiente.anterior = curr_node.siguiente
				curr_node = curr_node.siguiente

		self.__size += 1

	def transversal(self):
		curr_node = self.__ref.siguiente
		if self.__ref != None:
			if self.__ref.siguiente != None:
				for i in range(self.__size):
					print( "<--",curr_node.dato, "-->" , end="")
					curr_node = curr_node.siguiente
				print('\n')
			else:
				print(self.__ref.dato, "-->" , end="")
				print('\n')
		else:
			print("Lista vacia")
	def reverse_transversal(self):
		curr_node = self.__ref
		if self.__ref != None:
			if self.__ref.siguiente != None:
				for i in range(self.__size):
					print( "<--",curr_node.dato, "-->" , end="")
					curr_node = curr_node.anterior
				print('\n')
			else:
				print(self.__ref.dato, "-->" , end="")
				print('\n')
		else:
			print("Lista vacia")

	def search(self, value ):
		curr_node = self.__ref
		if self.__ref != None:
			for i in range(self.__size):
				if curr_node.dato == value:
					return True
				curr_node = curr_node.siguiente
		return False

	def remove(self,value):

		curr_node = self.__ref
		if self.__ref != None:
			for i in range(self.__size):
				if curr_node.siguiente.dato == value:
					curr_node.siguiente = curr_node.siguiente.siguiente
					curr_node.siguiente.anterior = curr_node
					self.__size -= 1
					return None
				else:
					curr_node = curr_node.siguiente



x = DoubleCircularList()

x.insert(1)
x.insert(2)
x.insert(5)
x.remove(2)

x.transversal()
x.reverse_transversal()

x.insert(1)
x.insert(3)
x.insert(4)

x.transversal()
x.reverse_transversal()
```
