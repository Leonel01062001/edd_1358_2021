```
class Nodo:
    def __init__( self , dato, siguiente=None ):
        self.dato = dato
        self.siguiente = siguiente

class LinkedList:
  def __init__ ( self ):
    self.__head = None

  def is_empty( self ):
    return self.__head == None
  
  def tail( self ):
    curr_node = self.__head
    while curr_node != None and curr_node.siguiente != None:
      curr_node = curr_node.siguiente
    return curr_node

  def append( self, value ):
    nuevo = Nodo( value )
    if self.__head == None:
      self.__head = nuevo
    else:
      self.tail().siguiente = nuevo
  
  def transeversal(self ):
    curr_node = self.__head
    if curr_node == None:
      print("Lista vacia")
    else:
      while curr_node != None:
        print(f"{ curr_node.dato } --->", end ="")
        curr_node = curr_node.siguiente
    print( "\n" )

  def prepend( self, value ):
    nuevo = Nodo( value )
    
    if self.__head == None:
      self.__head == nuevo
    else:
      x = self.__head
      self.__head = nuevo
      nuevo.siguiente = x

  def remove ( self, value ):
    curr_node = self.__head
    if self.__head == value:
      self.__head = self.__head.siguiente
    else:
      while curr_node.siguiente != None:
        if curr_node.siguiente.dato == value:
          curr_node.siguiente = curr_node.siguiente.siguiente
          break
        else:
          curr_node = curr_node.siguiente

  def pop( self, posicion=None ):
    if posicion == 0:
      Aux = self.__head
      self.__head = self.__head.siguiente
      print( Aux.dato )
    contador = 0
    curr_node = self.__head

    while curr_node.siguiente != None:
      if contador+1 == posicion:
        aux = curr_node.siguiente.dato
        curr_node.siguiente = curr_node.siguiente.siguiente
        print( aux )
        break
      else:
        contador = contador + 1
        curr_node = curr_node.siguiente

    while curr_node.siguiente != None:
      if posicion < 0:
        print( "No existe ese elemento" )
        
      else:
        contador = contador + 1
        curr_node = curr_node.siguiente

  def Add_before(self, value, reference):
    curr_node = self.__head
    if self.__head == reference:
      nuevo = Nodo(value)
      nuevo.siguiente = self.__head
    else:
      while curr_node.siguiente != None:
        if curr_node.siguiente.dato == reference:
          nuevo = Nodo(value)
          nuevo.siguiente = curr_node.siguiente
          curr_node.siguiente = nuevo
          break
        else:
          curr_node = curr_node.siguiente

  def Add_after(self, value, reference):
    curr_node = self.__head
    if self.__head == reference:
      nuevo = Nodo(value)
      nuevo.siguiente = self.__head
    else:
      while curr_node.siguiente != None:
        if curr_node.siguiente.dato == reference:
          nuevo = Nodo(value)
          nuevo.siguiente = curr_node.siguiente.siguiente
          curr_node.siguiente.siguiente = nuevo
          break
        else:
          curr_node = curr_node.siguiente
        

  #def prepend(self, value):
   #     nuevo = Nodo(value, self.__head)
    #    self.__head = nuevo
```
